```python
# File path
file_path = '../data/raw/PET_PRI_SPT_S1_M.xls'

# Map of sheets to fuel types (skip 'Contents')
sheet_map = {
    "Data 1": "Crude Oil",
    "Data 2": "Conventional Gasoline",
    "Data 3": "RBOB Regular Gasoline",
    "Data 4": "No. 2 Heating Oil",
    "Data 5": "ULS Diesel",
    "Data 6": "Jet Fuel",
    "Data 7": "Propane"
}

# Create a list to hold all the cleaned DataFrames
combined_data = []

# Loop through and collect each sheet
for sheet, fuel_name in sheet_map.items():
    df = pd.read_excel(file_path, sheet_name=sheet, skiprows=3)

    # Assume first column is date, second is price
    df = df.iloc[:, :2].copy()
    df.columns = ['Date', 'Price']

    # Convert to datetime
    df['Date'] = pd.to_datetime(df['Date'], format='%b-%Y', errors='coerce')

    # Drop rows with missing values
    df = df.dropna(subset=['Date', 'Price'])

    # Add a fuel type column
    df['Fuel Type'] = fuel_name

    # Append to list
    combined_data.append(df)

# Concatenate all data
all_fuels_df = pd.concat(combined_data, ignore_index=True)

# Export to CSV
all_fuels_df.to_csv('../data/cleaned/all_fuel_prices.csv', index=False)
```
